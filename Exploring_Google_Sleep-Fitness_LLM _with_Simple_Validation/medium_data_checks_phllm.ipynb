```python
# Step 1: Reload the data
import json
import pandas as pd
import numpy as np
import re
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

def load_jsonl(file_path):
    """Load JSONL file"""
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    return data

# Load both datasets
sleep_data = load_jsonl('consumer-health-research/phllm/data/sleep_case_studies.all.jsonl')
fitness_data = load_jsonl('consumer-health-research/phllm/data/fitness_case_studies.all.jsonl')

print(f"üìä Reloaded: {len(sleep_data)} sleep cases, {len(fitness_data)} fitness cases")
```

```python
# Step 2: Now define the PHLLMDataset class and initialize it
class PHLLMDataset:
    def __init__(self, sleep_data, fitness_data):
        self.sleep_data = sleep_data
        self.fitness_data = fitness_data

    def get_split_data(self, split_type="test"):
        """Get data for specific split (train/validation/test/holdout)"""
        sleep_split = [case for case in self.sleep_data if case.get('split') == split_type]
        fitness_split = [case for case in self.fitness_data if case.get('split') == split_type]
        return sleep_split, fitness_split

    def parse_sleep_input(self, sleep_case):
        """Parse structured sleep input data"""
        input_text = sleep_case['input']

        # Extract demographics
        lines = input_text.split('\n')
        demo_line = lines[0] if lines else ""

        # Extract age and gender using regex
        age_match = re.search(r'\[(\d+)-(\d+)\]', demo_line)
        gender_match = re.search(r'(male|female)', demo_line.lower())

        age_range = age_match.groups() if age_match else (None, None)
        gender = gender_match.group(1) if gender_match else None

        # Parse sleep data table
        sleep_logs = []
        in_data_section = False

        for line in lines:
            if "Day of Week" in line and "Sleep Score" in line:
                in_data_section = True
                continue

            if in_data_section and line.strip():
                # Parse sleep log line
                parts = line.split()
                if len(parts) >= 10:  # Ensure we have enough data
                    try:
                        log_entry = {
                            'day': parts[0] + " " + parts[1],
                            'day_of_week': parts[2],
                            'sleep_score': float(parts[3]) if parts[3] != '-' else None,
                            'sleep_duration': parts[7] if len(parts) > 7 else None,
                            'efficiency': float(parts[10]) if len(parts) > 10 and parts[10] != '-' else None,
                        }
                        sleep_logs.append(log_entry)
                    except (ValueError, IndexError):
                        continue

        return {
            'age_range': age_range,
            'gender': gender,
            'sleep_logs': sleep_logs,
            'demographics_text': demo_line
        }

    def parse_fitness_input(self, fitness_case):
        """Parse structured fitness input data"""
        return {
            'demographics': fitness_case.get('demographics_input', ''),
            'training_load': fitness_case.get('training_load_input', ''),
            'sleep': fitness_case.get('sleep_input', ''),
            'health_metrics': fitness_case.get('health_metrics_input', ''),
            'subjective_readiness': fitness_case.get('subjective_readiness_input', ''),
            'muscle_soreness': fitness_case.get('muscle_soreness_input', '')
        }

# Initialize the dataset
dataset = PHLLMDataset(sleep_data, fitness_data)

# Get test data for evaluation
test_sleep, test_fitness = dataset.get_split_data("test")
print(f"üìä Test Data: {len(test_sleep)} sleep cases, {len(test_fitness)} fitness cases")

# Parse a sample case
if test_sleep:
    sample_parsed = dataset.parse_sleep_input(test_sleep[0])
    print(f"üìã Sample parsed sleep case:")
    print(f"  Gender: {sample_parsed['gender']}")
    print(f"  Age range: {sample_parsed['age_range']}")
    print(f"  Number of sleep logs: {len(sample_parsed['sleep_logs'])}")
    if sample_parsed['sleep_logs']:
        print(f"  Sample log: {sample_parsed['sleep_logs'][0]}")
```

```python
# Step 3: Define the evaluator class
class PHLLMEvaluator:
    def __init__(self):
        # Evaluation rubric based on the paper (15 criteria, 1-5 scale)
        self.rubric_criteria = [
            "uses_important_domain_knowledge",
            "provides_important_interpretations",
            "references_important_user_data",
            "no_incorrect_domain_knowledge",
            "no_unimportant_interpretations",
            "no_incorrect_user_data_references",
            "appropriate_personalization",
            "no_confabulations",
            "no_unwarranted_assumptions",
            "low_potential_for_harm",
            "good_readability",
            "appropriate_length",
            "clear_structure",
            "actionable_recommendations",
            "overall_quality"
        ]

    def create_case_study_prompt(self, case_data, domain="sleep"):
        """Create standardized prompts for testing models"""

        if domain == "sleep":
            parsed = dataset.parse_sleep_input(case_data)

            prompt = f"""You are a sleep medicine expert analyzing wearable device data.

Patient Information:
- Gender: {parsed['gender']}
- Age: {parsed['age_range'][0]}-{parsed['age_range'][1]} years old

Sleep Data ({len(parsed['sleep_logs'])} days of data):
{case_data['input']}

Please provide a comprehensive analysis with:

1. **Insights**: Key observations about this person's sleep patterns
2. **Etiology**: Possible underlying causes or factors contributing to observed patterns
3. **Recommendations**: Specific, actionable advice for improving sleep

Format your response with clear sections and be specific about the data you're referencing."""

        else:  # fitness
            inputs = dataset.parse_fitness_input(case_data)

            prompt = f"""You are a fitness expert analyzing training and health data.

{inputs['demographics']}

Training Load Data:
{inputs['training_load'][:500]}...

Sleep Data:
{inputs['sleep'][:300]}...

Health Metrics:
{inputs['health_metrics'][:300]}...

Subjective Assessment:
{inputs['subjective_readiness']}

Please provide analysis for:
1. **Demographics**: Considerations based on age, BMI, etc.
2. **Training Load**: Analysis of recent activity patterns
3. **Sleep**: Sleep quality and recovery assessment
4. **Health Metrics**: Physiological indicators analysis
5. **Readiness Assessment**: Overall workout readiness and recommendations"""

        return prompt

# Initialize evaluator
evaluator = PHLLMEvaluator()

# Test the prompt creation
if test_sleep:
    sample_prompt = evaluator.create_case_study_prompt(test_sleep[0], "sleep")
    print("üìù Sample Sleep Case Prompt:")
    print(sample_prompt[:800] + "...")

    print(f"\nü©∫ Expert Response for comparison:")
    print(f"Insights: {test_sleep[0]['insight_output'][:200]}...")
    print(f"Etiology: {test_sleep[0]['etiology_output'][:200]}...")
    print(f"Recommendations: {test_sleep[0]['recommendation_output'][:200]}...")
```

```python
# Let's create a comprehensive testing system
class ModelComparison:
    def __init__(self, dataset, evaluator):
        self.dataset = dataset
        self.evaluator = evaluator
        self.results = []

    def test_with_simulated_model(self, case_data, domain="sleep"):
        """Test with a simulated model response (for demonstration)"""

        if domain == "sleep":
            # Create a simple rule-based response for testing
            parsed = self.dataset.parse_sleep_input(case_data)

            # Extract some basic stats from the sleep logs
            sleep_scores = [log.get('sleep_score') for log in parsed['sleep_logs'] if log.get('sleep_score')]
            avg_score = np.mean(sleep_scores) if sleep_scores else 0

            simulated_response = f"""**Insights:**
Based on the {len(parsed['sleep_logs'])} days of sleep data for this {parsed['gender']} aged {parsed['age_range'][0]}-{parsed['age_range'][1]}, the average sleep score is {avg_score:.1f}.

**Etiology:**
The sleep patterns show variability that may be related to age-related changes in sleep architecture and circadian rhythm regulation.

**Recommendations:**
1. Maintain consistent sleep schedule
2. Consider sleep hygiene improvements
3. Monitor for age-related sleep changes"""

        else:  # fitness
            simulated_response = """**Training Load Analysis:**
Recent activity shows moderate training load with room for optimization.

**Sleep Assessment:**
Sleep patterns indicate adequate recovery potential.

**Health Metrics:**
Physiological indicators suggest good baseline fitness.

**Readiness Assessment:**
Overall readiness appears moderate. Consider gradual intensity increases."""

        return simulated_response

    def evaluate_response_quality(self, model_response, expert_response, case_data):
        """Evaluate response quality using multiple metrics"""

        metrics = {}

        # Length comparison
        metrics['model_length'] = len(model_response)
        metrics['expert_length'] = len(expert_response)
        metrics['length_ratio'] = metrics['model_length'] / max(metrics['expert_length'], 1)

        # Structure analysis
        model_sections = len([line for line in model_response.split('\n') if line.strip().startswith('**')])
        expert_sections = len([line for line in expert_response.split('\n') if line.strip().startswith('**')])
        metrics['model_sections'] = model_sections
        metrics['expert_sections'] = expert_sections

        # Content analysis (basic)
        metrics['mentions_data'] = model_response.lower().count('data') + model_response.lower().count('day')
        metrics['mentions_sleep'] = model_response.lower().count('sleep')
        metrics['mentions_recommendation'] = model_response.lower().count('recommend')

        # Actionability score (count specific suggestions)
        action_words = ['consider', 'try', 'avoid', 'maintain', 'increase', 'decrease', 'limit']
        metrics['actionability_score'] = sum(model_response.lower().count(word) for word in action_words)

        return metrics

    def run_comparison_study(self, test_cases, domain="sleep"):
        """Run a complete comparison study"""

        results = []

        for i, case in enumerate(test_cases[:5]):  # Test first 5 cases
            print(f"üìä Processing case {i+1}/{min(5, len(test_cases))}: {case['case_study_id']}")

            # Create prompt
            prompt = self.evaluator.create_case_study_prompt(case, domain)

            # Get simulated model response
            model_response = self.test_with_simulated_model(case, domain)

            # Get expert response
            if domain == "sleep":
                expert_response = f"**Insights:**\n{case['insight_output']}\n\n**Etiology:**\n{case['etiology_output']}\n\n**Recommendations:**\n{case['recommendation_output']}"
            else:
                expert_response = case.get('readiness_assessment_output', '')

            # Evaluate quality
            metrics = self.evaluate_response_quality(model_response, expert_response, case)

            # Store results
            result = {
                'case_id': case['case_study_id'],
                'domain': domain,
                'prompt_length': len(prompt),
                'model_response': model_response,
                'expert_response': expert_response,
                'metrics': metrics,
                'case_data': case
            }
            results.append(result)

        return results

# Initialize comparison system
comparison = ModelComparison(dataset, evaluator)

# Run comparison study on sleep data
sleep_results = comparison.run_comparison_study(test_sleep, "sleep")
print(f"‚úÖ Completed sleep analysis on {len(sleep_results)} cases")
```

```python
# Statistical comparison following the paper's methodology
def statistical_analysis(results, domain="sleep"):
    """Perform statistical analysis similar to the paper"""

    print(f"üìä STATISTICAL ANALYSIS - {domain.upper()}")
    print("=" * 50)

    # Extract metrics for analysis
    model_lengths = [r['metrics']['model_length'] for r in results]
    expert_lengths = [r['metrics']['expert_length'] for r in results]
    length_ratios = [r['metrics']['length_ratio'] for r in results]
    actionability_scores = [r['metrics']['actionability_score'] for r in results]

    # Bootstrap confidence intervals (as used in the paper)
    def bootstrap_ci(data, n_bootstrap=1000, confidence=0.95):
        """Calculate bootstrap confidence interval"""
        bootstrap_means = []
        for _ in range(n_bootstrap):
            sample = np.random.choice(data, size=len(data), replace=True)
            bootstrap_means.append(np.mean(sample))

        alpha = 1 - confidence
        lower = np.percentile(bootstrap_means, 100 * alpha/2)
        upper = np.percentile(bootstrap_means, 100 * (1 - alpha/2))
        return lower, upper

    # Calculate statistics
    print("üìà Length Analysis:")
    model_mean = np.mean(model_lengths)
    expert_mean = np.mean(expert_lengths)
    model_ci = bootstrap_ci(model_lengths)
    expert_ci = bootstrap_ci(expert_lengths)

    print(f"Model length: {model_mean:.0f} ¬± {(model_ci[1]-model_ci[0])/2:.0f} chars (95% CI: {model_ci[0]:.0f}-{model_ci[1]:.0f})")
    print(f"Expert length: {expert_mean:.0f} ¬± {(expert_ci[1]-expert_ci[0])/2:.0f} chars (95% CI: {expert_ci[0]:.0f}-{expert_ci[1]:.0f})")

    # Wilcoxon rank-sum test (as used in the paper)
    from scipy import stats

    statistic, p_value = stats.ranksums(model_lengths, expert_lengths)
    print(f"Wilcoxon rank-sum test: statistic={statistic:.2f}, p-value={p_value:.4f}")

    if p_value < 0.05:
        print("‚ö†Ô∏è Significant difference in response lengths")
    else:
        print("‚úÖ No significant difference in response lengths")

    # Effect size calculation
    n_total = len(model_lengths) + len(expert_lengths)
    effect_size = abs(statistic) / np.sqrt(n_total)
    print(f"Effect size (r): {effect_size:.3f}")

    print(f"\nüìä Actionability Analysis:")
    action_mean = np.mean(actionability_scores)
    action_ci = bootstrap_ci(actionability_scores)
    print(f"Actionability score: {action_mean:.1f} ¬± {(action_ci[1]-action_ci[0])/2:.1f} (95% CI: {action_ci[0]:.1f}-{action_ci[1]:.1f})")

    return {
        'model_length_mean': model_mean,
        'expert_length_mean': expert_mean,
        'length_ratio_mean': np.mean(length_ratios),
        'actionability_mean': action_mean,
        'wilcoxon_statistic': statistic,
        'wilcoxon_p_value': p_value,
        'effect_size': effect_size
    }

# Run statistical analysis
stats_results = statistical_analysis(sleep_results, "sleep")
```

```python

# Implement the detailed evaluation rubric from the paper
class DetailedEvaluationRubric:
    def __init__(self):
        # 15 evaluation criteria from Supplementary Table 9 (based on paper description)
        self.criteria = {
            1: {
                "name": "uses_important_domain_knowledge",
                "description": "Uses important domain knowledge relevant to sleep/fitness",
                "scale": "1=No domain knowledge, 5=Excellent domain knowledge"
            },
            2: {
                "name": "provides_important_interpretations",
                "description": "Provides important interpretations of the data",
                "scale": "1=No interpretations, 5=Excellent interpretations"
            },
            3: {
                "name": "references_important_user_data",
                "description": "References important and relevant user data",
                "scale": "1=No data references, 5=Excellent data integration"
            },
            4: {
                "name": "no_incorrect_domain_knowledge",
                "description": "Avoids incorrect domain knowledge",
                "scale": "1=Many errors, 5=No domain errors"
            },
            5: {
                "name": "no_unimportant_interpretations",
                "description": "Avoids unimportant or irrelevant interpretations",
                "scale": "1=Many irrelevant points, 5=All relevant"
            },
            6: {
                "name": "no_incorrect_user_data_references",
                "description": "Avoids incorrect references to user data",
                "scale": "1=Many data errors, 5=No data errors"
            },
            7: {
                "name": "appropriate_personalization",
                "description": "Appropriately personalizes advice for the user",
                "scale": "1=Generic advice, 5=Highly personalized"
            },
            8: {
                "name": "no_confabulations",
                "description": "Avoids making up facts or data not present",
                "scale": "1=Many fabrications, 5=No fabrications"
            },
            9: {
                "name": "no_unwarranted_assumptions",
                "description": "Avoids unwarranted assumptions about the user",
                "scale": "1=Many assumptions, 5=No assumptions"
            },
            10: {
                "name": "low_potential_for_harm",
                "description": "Low potential for causing harm with advice",
                "scale": "1=High harm potential, 5=Very safe advice"
            },
            11: {
                "name": "good_readability",
                "description": "Text is clear, well-structured, and readable",
                "scale": "1=Poor readability, 5=Excellent readability"
            },
            12: {
                "name": "appropriate_length",
                "description": "Response length is appropriate for the task",
                "scale": "1=Too short/long, 5=Perfect length"
            },
            13: {
                "name": "clear_structure",
                "description": "Response has clear structure and organization",
                "scale": "1=No structure, 5=Excellent structure"
            },
            14: {
                "name": "actionable_recommendations",
                "description": "Provides specific, actionable recommendations",
                "scale": "1=Vague advice, 5=Highly actionable"
            },
            15: {
                "name": "overall_quality",
                "description": "Overall quality of the response",
                "scale": "1=Very poor, 5=Excellent"
            }
        }

    def automated_evaluation(self, response, case_data, expert_response, domain="sleep"):
        """
        Automated evaluation based on response analysis
        This approximates human expert evaluation
        """

        scores = {}

        # Parse the response for analysis
        response_lower = response.lower()
        response_lines = response.split('\n')

        # 1. Domain knowledge (check for medical/fitness terms)
        domain_terms = {
            'sleep': ['circadian', 'rem', 'deep sleep', 'sleep efficiency', 'chronotype', 'melatonin', 'sleep hygiene'],
            'fitness': ['trimp', 'heart rate', 'recovery', 'training load', 'acwr', 'vo2', 'lactate threshold']
        }

        relevant_terms = domain_terms.get(domain, [])
        domain_score = min(5, 1 + sum(1 for term in relevant_terms if term in response_lower))
        scores[1] = domain_score

        # 2. Interpretations (check for analytical language)
        interpretation_words = ['shows', 'indicates', 'suggests', 'pattern', 'trend', 'analysis']
        interpretation_count = sum(1 for word in interpretation_words if word in response_lower)
        scores[2] = min(5, 1 + interpretation_count)

        # 3. Data references (check for specific data mentions)
        data_indicators = ['average', 'score', 'duration', 'efficiency', 'day', 'hour', 'minute']
        data_references = sum(1 for indicator in data_indicators if indicator in response_lower)
        scores[3] = min(5, 1 + data_references // 2)

        # 4. No incorrect domain knowledge (high score unless obvious errors)
        # This is hard to automate, so we give benefit of doubt
        scores[4] = 4  # Assume mostly correct unless human review

        # 5. Relevant interpretations (penalize if too generic)
        generic_phrases = ['it is important', 'you should', 'it is recommended']
        generic_count = sum(1 for phrase in generic_phrases if phrase in response_lower)
        scores[5] = max(1, 5 - generic_count)

        # 6. Data accuracy (check for specific numbers/references)
        # Hard to automate, assume good unless obvious issues
        scores[6] = 4

        # 7. Personalization (check for personal pronouns and specific advice)
        personal_words = ['your', 'you', 'user', 'patient']
        personal_count = sum(1 for word in personal_words if word in response_lower)
        scores[7] = min(5, 1 + personal_count // 3)

        # 8. No confabulations (assume good unless review shows otherwise)
        scores[8] = 4

        # 9. No unwarranted assumptions (assume good)
        scores[9] = 4

        # 10. Safety (high score for health advice)
        harmful_words = ['ignore doctor', 'stop medication', 'dangerous']
        if any(word in response_lower for word in harmful_words):
            scores[10] = 2
        else:
            scores[10] = 5

        # 11. Readability (check structure and length)
        word_count = len(response.split())
        if 100 <= word_count <= 500:
            scores[11] = 5
        elif 50 <= word_count <= 800:
            scores[11] = 4
        else:
            scores[11] = 3

        # 12. Appropriate length (similar to readability)
        scores[12] = scores[11]

        # 13. Structure (check for headers and organization)
        structured_indicators = ['**', '##', '1.', '2.', '‚Ä¢', '-']
        structure_count = sum(1 for indicator in structured_indicators if indicator in response)
        scores[13] = min(5, 1 + structure_count)

        # 14. Actionable recommendations (check for specific actions)
        action_words = ['try', 'consider', 'avoid', 'maintain', 'increase', 'decrease', 'limit', 'aim for']
        action_count = sum(1 for word in action_words if word in response_lower)
        scores[14] = min(5, 1 + action_count)

        # 15. Overall quality (average of key metrics)
        key_scores = [scores[1], scores[2], scores[3], scores[7], scores[14]]
        scores[15] = round(np.mean(key_scores))

        return scores

    def calculate_aggregate_score(self, scores):
        """Calculate aggregate score from individual criteria"""
        return {
            'mean_score': np.mean(list(scores.values())),
            'total_score': sum(scores.values()),
            'max_possible': len(scores) * 5,
            'percentage': (sum(scores.values()) / (len(scores) * 5)) * 100
        }

    def compare_to_expert_baseline(self, model_scores, expert_baseline=4.75):
        """Compare model performance to expert baseline from paper"""
        model_mean = np.mean(list(model_scores.values()))

        return {
            'model_mean': model_mean,
            'expert_baseline': expert_baseline,
            'difference': model_mean - expert_baseline,
            'relative_performance': (model_mean / expert_baseline) * 100
        }

# Initialize detailed evaluator
detailed_evaluator = DetailedEvaluationRubric()

# Test the detailed evaluation
def test_detailed_evaluation():
    """Test the detailed evaluation system"""

    print("üß™ TESTING DETAILED EVALUATION RUBRIC")
    print("=" * 50)

    if not test_sleep:
        print("‚ùå No test data available")
        return

    # Get a test case
    test_case = test_sleep[0]

    # Create model response (using our simulated model)
    model_response = comparison.test_with_simulated_model(test_case, "sleep")

    # Get expert response
    expert_response = f"**Insights:**\n{test_case['insight_output']}\n\n**Etiology:**\n{test_case['etiology_output']}\n\n**Recommendations:**\n{test_case['recommendation_output']}"

    # Evaluate model response
    model_scores = detailed_evaluator.automated_evaluation(
        model_response, test_case, expert_response, "sleep"
    )

    # Calculate aggregate scores
    model_aggregate = detailed_evaluator.calculate_aggregate_score(model_scores)

    # Compare to expert baseline
    comparison_result = detailed_evaluator.compare_to_expert_baseline(model_scores)

    # Display results
    print(f"üìä DETAILED EVALUATION RESULTS")
    print(f"Case ID: {test_case['case_study_id']}")
    print()

    print("üìã Individual Criteria Scores (1-5 scale):")
    for criterion_id, score in model_scores.items():
        criterion_info = detailed_evaluator.criteria[criterion_id]
        print(f"  {criterion_id:2d}. {criterion_info['name']}: {score}/5")

    print(f"\nüìà Aggregate Scores:")
    print(f"  Mean score: {model_aggregate['mean_score']:.2f}/5")
    print(f"  Total score: {model_aggregate['total_score']}/{model_aggregate['max_possible']}")
    print(f"  Percentage: {model_aggregate['percentage']:.1f}%")

    print(f"\nüîÑ Comparison to Expert Baseline:")
    print(f"  Model mean: {comparison_result['model_mean']:.2f}")
    print(f"  Expert baseline: {comparison_result['expert_baseline']:.2f}")
    print(f"  Difference: {comparison_result['difference']:.2f}")
    print(f"  Relative performance: {comparison_result['relative_performance']:.1f}%")

    return model_scores, model_aggregate, comparison_result

# Run the detailed evaluation test
evaluation_results = test_detailed_evaluation()
```

```python
# new one
# Fixed Comprehensive comparison system
class ComprehensiveModelComparison:
    def __init__(self, dataset, evaluator, detailed_evaluator, real_tester):
        self.dataset = dataset
        self.evaluator = evaluator
        self.detailed_evaluator = detailed_evaluator
        self.real_tester = real_tester

    def run_full_comparison(self, test_cases, domain="sleep", api_keys=None):
        """Run comprehensive comparison across multiple models"""

        if not api_keys:
            api_keys = {}

        results = []

        for i, case in enumerate(test_cases[:3]):  # Test first 3 cases
            print(f"üîÑ Processing case {i+1}/3: {case['case_study_id']}")

            # Create prompt
            prompt = self.evaluator.create_case_study_prompt(case, domain)

            # Get expert response
            if domain == "sleep":
                expert_response = f"**Insights:**\n{case['insight_output']}\n\n**Etiology:**\n{case['etiology_output']}\n\n**Recommendations:**\n{case['recommendation_output']}"
            else:
                expert_response = case.get('readiness_assessment_output', '')

            # Test all available models
            model_responses = {}

            # Simulated model (baseline)
            model_responses['Simulated'] = comparison.test_with_simulated_model(case, domain)

            # Real models (if API keys provided)
            if api_keys.get('openai'):
                print("  ü§ñ Testing GPT-4...")
                gpt4_response = self.real_tester.test_openai_gpt4(prompt, api_keys['openai'])
                if not gpt4_response.startswith('‚ùå'):
                    model_responses['GPT-4'] = gpt4_response
                    print(f"  ‚úÖ GPT-4 response: {len(gpt4_response)} chars")
                else:
                    print(f"  ‚ùå GPT-4 error: {gpt4_response}")

            if api_keys.get('anthropic'):
                print("  ü§ñ Testing Claude...")
                claude_response = self.real_tester.test_anthropic_claude(prompt, api_keys['anthropic'])
                if not claude_response.startswith('‚ùå'):
                    model_responses['Claude'] = claude_response
                    print(f"  ‚úÖ Claude response: {len(claude_response)} chars")
                else:
                    print(f"  ‚ùå Claude error: {claude_response}")

            if api_keys.get('google'):
                print("  ü§ñ Testing Gemini...")
                gemini_response = self.real_tester.test_google_gemini(prompt, api_keys['google'])
                if not gemini_response.startswith('‚ùå'):
                    model_responses['Gemini'] = gemini_response
                    print(f"  ‚úÖ Gemini response: {len(gemini_response)} chars")
                else:
                    print(f"  ‚ùå Gemini error: {gemini_response}")

            # Evaluate each model
            case_results = {
                'case_id': case['case_study_id'],
                'domain': domain,
                'expert_response': expert_response,
                'model_evaluations': {}
            }

            for model_name, response in model_responses.items():
                print(f"  üìä Evaluating {model_name}...")

                # Detailed evaluation
                scores = self.detailed_evaluator.automated_evaluation(
                    response, case, expert_response, domain
                )
                aggregate = self.detailed_evaluator.calculate_aggregate_score(scores)

                case_results['model_evaluations'][model_name] = {
                    'response': response,
                    'scores': scores,
                    'aggregate': aggregate
                }

                print(f"     Score: {aggregate['mean_score']:.2f}/5")

            results.append(case_results)
            print()

        return results

    def create_comparison_report(self, comparison_results):
        """Create a comprehensive comparison report"""

        print("üìä COMPREHENSIVE MODEL COMPARISON REPORT")
        print("=" * 60)

        # Aggregate results across all cases
        model_performance = {}

        for case_result in comparison_results:
            for model_name, evaluation in case_result['model_evaluations'].items():
                if model_name not in model_performance:
                    model_performance[model_name] = {
                        'scores': [],
                        'mean_scores': [],
                        'case_count': 0
                    }

                model_performance[model_name]['mean_scores'].append(evaluation['aggregate']['mean_score'])
                model_performance[model_name]['case_count'] += 1

        # Calculate overall performance
        print("üèÜ OVERALL PERFORMANCE RANKING:")
        rankings = []
        for model_name, perf in model_performance.items():
            overall_mean = np.mean(perf['mean_scores'])
            rankings.append((model_name, overall_mean, perf['case_count']))

        rankings.sort(key=lambda x: x[1], reverse=True)

        for rank, (model, score, count) in enumerate(rankings, 1):
            print(f"  {rank}. {model}: {score:.2f}/5 (n={count} cases)")

        # Statistical comparison
        print(f"\nüìà STATISTICAL ANALYSIS:")
        expert_baseline = 4.75  # From paper
        ph_llm_baseline = 4.61  # PH-LLM performance from paper

        if rankings:
            best_model = rankings[0]
            print(f"Best performing model: {best_model[0]} ({best_model[1]:.2f}/5)")
            print(f"PH-LLM baseline: {ph_llm_baseline:.2f}/5")
            print(f"Expert baseline: {expert_baseline:.2f}/5")

            # Compare to PH-LLM
            gap_to_phllm = best_model[1] - ph_llm_baseline
            print(f"Gap to PH-LLM: {gap_to_phllm:+.2f}")

            if best_model[1] >= ph_llm_baseline:
                print("‚úÖ Model matches or exceeds PH-LLM performance!")
            elif best_model[1] >= ph_llm_baseline * 0.95:
                print("‚ö†Ô∏è Model approaches PH-LLM performance")
            else:
                print("‚ùå Significant gap compared to PH-LLM")

            # Compare to Expert
            gap_to_expert = expert_baseline - best_model[1]
            print(f"Gap to Expert: {gap_to_expert:.2f}")

        return model_performance, rankings

# Fixed example usage
def run_complete_evaluation_fixed():
    """Run the complete evaluation with real APIs - FIXED VERSION"""

    # üîë SET YOUR API KEYS HERE:
    api_keys = {
        #'openai': 'your_key_here',      # Replace with your actual key
        'anthropic': 'your_key_here',           # Add if you have it
        # 'google': 'AIza...'                  # Add if you have it
    }

    # Initialize comprehensive comparison
    comprehensive = ComprehensiveModelComparison(
        dataset, evaluator, detailed_evaluator, real_tester
    )

    # Run comparison
    print("üöÄ Starting FIXED comprehensive model comparison...")
    comparison_results = comprehensive.run_full_comparison(
        test_sleep, "sleep", api_keys
    )

    # Generate report
    performance, rankings = comprehensive.create_comparison_report(comparison_results)

    return comparison_results, performance, rankings

# Run the FIXED evaluation
evaluation_output_fixed = run_complete_evaluation_fixed()
```

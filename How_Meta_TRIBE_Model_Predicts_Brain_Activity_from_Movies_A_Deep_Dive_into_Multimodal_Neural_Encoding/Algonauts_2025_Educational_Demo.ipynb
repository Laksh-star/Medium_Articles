```python
# Algonauts 2025 Educational Demo - Complete Colab Notebook
# This notebook provides a working educational version of brain encoding concepts

# ================================
# PART 1: ENVIRONMENT SETUP
# ================================

import sys
print("üîß Setting up environment...")
print(f"Python version: {sys.version}")

# Install required packages with error handling
import subprocess
import importlib

def install_package(package):
    """Install package with error handling"""
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"‚úÖ Successfully installed {package}")
    except subprocess.CalledProcessError as e:
        print(f"‚ùå Failed to install {package}: {e}")

# Install packages one by one to catch errors
packages = [
    "torch torchvision torchaudio",
    "transformers==4.36.0",
    "scikit-learn",
    "matplotlib",
    "seaborn",
    "numpy",
    "pandas"
]

for package in packages:
    install_package(package)

# ================================
# PART 2: CHECK ENVIRONMENT
# ================================

print("\nüîç Checking environment...")

try:
    import torch
    import torch.nn as nn
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    from sklearn.decomposition import PCA
    from sklearn.metrics import r2_score, mean_squared_error
    from sklearn.model_selection import train_test_split

    print(f"‚úÖ PyTorch version: {torch.__version__}")
    print(f"‚úÖ CUDA available: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"‚úÖ CUDA version: {torch.version.cuda}")
        print(f"‚úÖ GPU device: {torch.cuda.get_device_name(0)}")
    else:
        print("‚ö†Ô∏è CUDA not available, using CPU")

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"‚úÖ Using device: {device}")

except ImportError as e:
    print(f"‚ùå Import error: {e}")
    print("Please restart runtime and try again")

# ================================
# PART 3: SIMPLIFIED TRIBE MODEL
# ================================

print("\nüß† Building Simplified TRIBE Architecture...")

class SimplifiedTrimodalEncoder(nn.Module):
    """
    Educational implementation of TRIBE (Trimodal Brain Encoder)

    This is a simplified version for educational purposes.
    The real TRIBE model has 1B parameters and uses:
    - V-JEPA 2 for video encoding
    - Wav2Vec2-BERT for audio encoding
    - Llama 3.2 for text encoding
    """

    def __init__(self,
                 visual_dim=512,
                 audio_dim=256,
                 text_dim=768,
                 brain_regions=1000,
                 hidden_dim=256):
        super().__init__()

        # Modality-specific encoders (simplified versions)
        self.visual_encoder = nn.Sequential(
            nn.Linear(visual_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU()
        )

        self.audio_encoder = nn.Sequential(
            nn.Linear(audio_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, hidden_dim // 4),
            nn.ReLU()
        )

        self.text_encoder = nn.Sequential(
            nn.Linear(text_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU()
        )

        # Attention mechanism for modality fusion
        self.attention = nn.MultiheadAttention(
            embed_dim=hidden_dim // 2,
            num_heads=8,
            batch_first=True
        )

        # Fusion and brain prediction network
        combined_dim = 128 + 64  # attended_vt (128) + a_encoded (64) = 192

        self.fusion = nn.Sequential(
            nn.Linear(combined_dim, hidden_dim * 2),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_dim * 2, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, brain_regions)
        )

        # Store dimensions for reference
        self.visual_dim = visual_dim
        self.audio_dim = audio_dim
        self.text_dim = text_dim
        self.brain_regions = brain_regions

    def forward(self, visual_features, audio_features, text_features):
        """
        Forward pass through the trimodal encoder

        Args:
            visual_features: Shape [batch_size, visual_dim]
            audio_features: Shape [batch_size, audio_dim]
            text_features: Shape [batch_size, text_dim]

        Returns:
            brain_responses: Shape [batch_size, brain_regions]
        """
        batch_size = visual_features.shape[0]

        # Encode each modality
        v_encoded = self.visual_encoder(visual_features)  # [batch, 128]
        a_encoded = self.audio_encoder(audio_features)    # [batch, 64]
        t_encoded = self.text_encoder(text_features)      # [batch, 128]

        # Apply attention to visual and text features (most important modalities)
        vt_features = torch.stack([v_encoded, t_encoded], dim=1)  # [batch, 2, 128]
        attended_features, _ = self.attention(vt_features, vt_features, vt_features)
        attended_vt = attended_features.mean(dim=1)  # [batch, 128]

        # Combine all modalities
        combined = torch.cat([attended_vt, a_encoded], dim=-1)  # [batch, 128+64=192]

        # Predict brain responses
        brain_responses = self.fusion(combined)

        return brain_responses

    def get_model_info(self):
        """Return model information"""
        total_params = sum(p.numel() for p in self.parameters())
        trainable_params = sum(p.numel() for p in self.parameters() if p.requires_grad)

        return {
            'total_parameters': total_params,
            'trainable_parameters': trainable_params,
            'visual_dim': self.visual_dim,
            'audio_dim': self.audio_dim,
            'text_dim': self.text_dim,
            'brain_regions': self.brain_regions
        }

# Initialize the model
print("üèóÔ∏è Initializing model...")
model = SimplifiedTrimodalEncoder().to(device)
model_info = model.get_model_info()

print(f"‚úÖ Model created successfully!")
print(f"üìä Total parameters: {model_info['total_parameters']:,}")
print(f"üìä Trainable parameters: {model_info['trainable_parameters']:,}")

# ================================
# PART 4: SYNTHETIC DATA GENERATION
# ================================

print("\nüé≤ Generating synthetic multimodal data...")

def generate_synthetic_data(n_samples=1000, noise_level=0.1):
    """
    Generate synthetic multimodal data that simulates:
    - Visual features from video frames
    - Audio features from soundtracks
    - Text features from dialogue/captions
    - Corresponding brain responses
    """

    # Set random seed for reproducibility
    torch.manual_seed(42)
    np.random.seed(42)

    # Generate base patterns that will influence brain responses
    n_patterns = 20
    base_patterns = torch.randn(n_patterns, model_info['brain_regions'])

    # Generate multimodal features with some correlation structure
    visual_features = torch.randn(n_samples, model_info['visual_dim'])
    audio_features = torch.randn(n_samples, model_info['audio_dim'])
    text_features = torch.randn(n_samples, model_info['text_dim'])

    # Create realistic brain responses based on feature combinations
    brain_responses = torch.zeros(n_samples, model_info['brain_regions'])

    for i in range(n_samples):
        # Select random patterns based on feature content
        visual_weight = torch.softmax(visual_features[i][:n_patterns], dim=0)
        audio_weight = torch.softmax(audio_features[i][:n_patterns], dim=0)
        text_weight = torch.softmax(text_features[i][:n_patterns], dim=0)

        # Combine patterns with modality-specific weights
        combined_weight = 0.5 * visual_weight + 0.3 * text_weight + 0.2 * audio_weight

        # Generate brain response as weighted combination of patterns
        brain_response = torch.sum(
            combined_weight.unsqueeze(1) * base_patterns,
            dim=0
        )

        # Add noise
        brain_response += noise_level * torch.randn_like(brain_response)
        brain_responses[i] = brain_response

    return visual_features, audio_features, text_features, brain_responses

# Generate dataset
print("üéØ Creating synthetic brain-behavior dataset...")
visual_data, audio_data, text_data, brain_data = generate_synthetic_data(n_samples=2000)

print(f"‚úÖ Generated dataset:")
print(f"   Visual features: {visual_data.shape}")
print(f"   Audio features: {audio_data.shape}")
print(f"   Text features: {text_data.shape}")
print(f"   Brain responses: {brain_data.shape}")

# ================================
# PART 5: TRAINING SETUP
# ================================

print("\nüöÄ Setting up training...")

# Split data into train/validation/test sets
def create_data_splits(visual, audio, text, brain, test_size=0.2, val_size=0.2):
    """Create train/validation/test splits"""

    # First split: separate test set
    indices = np.arange(len(visual))
    train_val_idx, test_idx = train_test_split(
        indices, test_size=test_size, random_state=42
    )

    # Second split: separate validation from training
    train_idx, val_idx = train_test_split(
        train_val_idx, test_size=val_size/(1-test_size), random_state=42
    )

    # Create splits
    splits = {}
    for split_name, idx in [('train', train_idx), ('val', val_idx), ('test', test_idx)]:
        splits[split_name] = {
            'visual': visual[idx],
            'audio': audio[idx],
            'text': text[idx],
            'brain': brain[idx]
        }

    return splits

# Create data splits
data_splits = create_data_splits(visual_data, audio_data, text_data, brain_data)

print(f"üìä Data splits created:")
for split_name, split_data in data_splits.items():
    print(f"   {split_name}: {split_data['visual'].shape[0]} samples")

# Create data loaders
from torch.utils.data import Dataset, DataLoader

class BrainDataset(Dataset):
    """Dataset class for multimodal brain data"""

    def __init__(self, visual, audio, text, brain):
        self.visual = visual.float()
        self.audio = audio.float()
        self.text = text.float()
        self.brain = brain.float()

    def __len__(self):
        return len(self.visual)

    def __getitem__(self, idx):
        return {
            'visual': self.visual[idx],
            'audio': self.audio[idx],
            'text': self.text[idx],
            'brain': self.brain[idx]
        }

# Create datasets and data loaders
batch_size = 32

datasets = {}
dataloaders = {}

for split_name, split_data in data_splits.items():
    datasets[split_name] = BrainDataset(
        split_data['visual'],
        split_data['audio'],
        split_data['text'],
        split_data['brain']
    )

    shuffle = (split_name == 'train')
    dataloaders[split_name] = DataLoader(
        datasets[split_name],
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=False
    )

print(f"‚úÖ Data loaders created with batch size: {batch_size}")

# ================================
# PART 6: TRAINING LOOP
# ================================

print("\nüéØ Training the model...")

# Training configuration
learning_rate = 1e-3
num_epochs = 50
patience = 10  # For early stopping

# Loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=5, verbose=True
)

# Training history
history = {
    'train_loss': [],
    'val_loss': [],
    'train_r2': [],
    'val_r2': []
}

def compute_r2_score(y_true, y_pred):
    """Compute R¬≤ score for brain predictions"""
    y_true_np = y_true.detach().cpu().numpy()
    y_pred_np = y_pred.detach().cpu().numpy()

    # Compute R¬≤ for each brain region and take the mean
    r2_scores = []
    for i in range(y_true_np.shape[1]):
        r2 = r2_score(y_true_np[:, i], y_pred_np[:, i])
        if not np.isnan(r2):  # Skip NaN values
            r2_scores.append(r2)

    return np.mean(r2_scores) if r2_scores else 0.0

def train_epoch(model, dataloader, criterion, optimizer):
    """Train for one epoch"""
    model.train()
    total_loss = 0
    all_predictions = []
    all_targets = []

    for batch in dataloader:
        # Move batch to device
        visual = batch['visual'].to(device)
        audio = batch['audio'].to(device)
        text = batch['text'].to(device)
        brain = batch['brain'].to(device)

        # Forward pass
        optimizer.zero_grad()
        predictions = model(visual, audio, text)
        loss = criterion(predictions, brain)

        # Backward pass
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        # Track metrics
        total_loss += loss.item()
        all_predictions.append(predictions)
        all_targets.append(brain)

    # Compute average loss and R¬≤
    avg_loss = total_loss / len(dataloader)
    all_predictions = torch.cat(all_predictions, dim=0)
    all_targets = torch.cat(all_targets, dim=0)
    avg_r2 = compute_r2_score(all_targets, all_predictions)

    return avg_loss, avg_r2

def validate_epoch(model, dataloader, criterion):
    """Validate for one epoch"""
    model.eval()
    total_loss = 0
    all_predictions = []
    all_targets = []

    with torch.no_grad():
        for batch in dataloader:
            # Move batch to device
            visual = batch['visual'].to(device)
            audio = batch['audio'].to(device)
            text = batch['text'].to(device)
            brain = batch['brain'].to(device)

            # Forward pass
            predictions = model(visual, audio, text)
            loss = criterion(predictions, brain)

            # Track metrics
            total_loss += loss.item()
            all_predictions.append(predictions)
            all_targets.append(brain)

    # Compute average loss and R¬≤
    avg_loss = total_loss / len(dataloader)
    all_predictions = torch.cat(all_predictions, dim=0)
    all_targets = torch.cat(all_targets, dim=0)
    avg_r2 = compute_r2_score(all_targets, all_predictions)

    return avg_loss, avg_r2

# Training loop with progress tracking
print("üîÑ Starting training loop...")
best_val_loss = float('inf')
patience_counter = 0

for epoch in range(num_epochs):
    # Training phase
    train_loss, train_r2 = train_epoch(model, dataloaders['train'], criterion, optimizer)

    # Validation phase
    val_loss, val_r2 = validate_epoch(model, dataloaders['val'], criterion)

    # Update learning rate scheduler
    scheduler.step(val_loss)

    # Track history
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_r2'].append(train_r2)
    history['val_r2'].append(val_r2)

    # Print progress
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f"Epoch {epoch+1:2d}/{num_epochs} | "
              f"Train Loss: {train_loss:.4f} | "
              f"Val Loss: {val_loss:.4f} | "
              f"Train R¬≤: {train_r2:.4f} | "
              f"Val R¬≤: {val_r2:.4f}")

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        # Save best model
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

print("‚úÖ Training completed!")

# Load best model
model.load_state_dict(torch.load('best_model.pth'))
print("‚úÖ Best model loaded")

# ================================
# PART 7: EVALUATION AND TESTING
# ================================

print("\nüìä Evaluating model performance...")

# Test the model
test_loss, test_r2 = validate_epoch(model, dataloaders['test'], criterion)
print(f"üéØ Test Results:")
print(f"   Test Loss: {test_loss:.4f}")
print(f"   Test R¬≤: {test_r2:.4f}")

# Get predictions for visualization
model.eval()
with torch.no_grad():
    # Get a batch from test set
    test_batch = next(iter(dataloaders['test']))
    visual_test = test_batch['visual'].to(device)
    audio_test = test_batch['audio'].to(device)
    text_test = test_batch['text'].to(device)
    brain_test = test_batch['brain'].to(device)

    # Make predictions
    predictions = model(visual_test, audio_test, text_test)

# Move to CPU for plotting
predictions_np = predictions.cpu().numpy()
targets_np = brain_test.cpu().numpy()

# ================================
# PART 8: VISUALIZATION
# ================================

print("\nüìà Creating visualizations...")

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

# Create comprehensive visualization
fig = plt.figure(figsize=(20, 15))

# 1. Training history
ax1 = plt.subplot(3, 4, 1)
epochs = range(1, len(history['train_loss']) + 1)
plt.plot(epochs, history['train_loss'], 'b-', label='Training Loss', linewidth=2)
plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training History - Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# 2. R¬≤ score history
ax2 = plt.subplot(3, 4, 2)
plt.plot(epochs, history['train_r2'], 'b-', label='Training R¬≤', linewidth=2)
plt.plot(epochs, history['val_r2'], 'r-', label='Validation R¬≤', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('R¬≤ Score')
plt.title('Training History - R¬≤ Score')
plt.legend()
plt.grid(True, alpha=0.3)

# 3. Prediction vs Target scatter plot
ax3 = plt.subplot(3, 4, 3)
sample_idx = 0
plt.scatter(targets_np[sample_idx], predictions_np[sample_idx], alpha=0.6, s=20)
plt.plot([targets_np[sample_idx].min(), targets_np[sample_idx].max()],
         [targets_np[sample_idx].min(), targets_np[sample_idx].max()], 'r--', linewidth=2)
plt.xlabel('Actual Brain Response')
plt.ylabel('Predicted Brain Response')
plt.title(f'Prediction vs Target (Sample {sample_idx})')
plt.grid(True, alpha=0.3)

# 4. Brain response patterns
ax4 = plt.subplot(3, 4, 4)
sample_response = predictions_np[0]
plt.plot(sample_response[:200], linewidth=1.5, label='Predicted')
plt.plot(targets_np[0][:200], linewidth=1.5, label='Actual', alpha=0.7)
plt.xlabel('Brain Region')
plt.ylabel('Response Magnitude')
plt.title('Brain Response Pattern (First 200 Regions)')
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Response distribution
ax5 = plt.subplot(3, 4, 5)
plt.hist(predictions_np.flatten(), bins=50, alpha=0.7, label='Predicted', density=True)
plt.hist(targets_np.flatten(), bins=50, alpha=0.7, label='Actual', density=True)
plt.xlabel('Response Magnitude')
plt.ylabel('Density')
plt.title('Distribution of Brain Responses')
plt.legend()
plt.grid(True, alpha=0.3)

# 6. PCA visualization
ax6 = plt.subplot(3, 4, 6)
pca = PCA(n_components=2)
pred_pca = pca.fit_transform(predictions_np)
target_pca = pca.transform(targets_np)
plt.scatter(pred_pca[:, 0], pred_pca[:, 1], alpha=0.6, label='Predicted', s=30)
plt.scatter(target_pca[:, 0], target_pca[:, 1], alpha=0.6, label='Actual', s=30)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
plt.title('PCA of Brain Responses')
plt.legend()
plt.grid(True, alpha=0.3)

# 7. Regional correlation heatmap
ax7 = plt.subplot(3, 4, 7)
# Compute correlation for first 50 regions (for visibility)
n_regions = 50
correlations = []
for i in range(n_regions):
    corr = np.corrcoef(targets_np[:, i], predictions_np[:, i])[0, 1]
    correlations.append(corr if not np.isnan(corr) else 0)

regions = np.arange(n_regions)
colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(correlations)))
bars = plt.bar(regions, correlations, color=colors)
plt.xlabel('Brain Region')
plt.ylabel('Correlation')
plt.title(f'Per-Region Correlation (First {n_regions} regions)')
plt.grid(True, alpha=0.3)

# 8. Model architecture diagram (text-based)
ax8 = plt.subplot(3, 4, 8)
ax8.text(0.1, 0.9, 'TRIBE Architecture', fontsize=16, fontweight='bold')
ax8.text(0.1, 0.8, f'Visual Encoder: {model_info["visual_dim"]} ‚Üí 128', fontsize=10)
ax8.text(0.1, 0.7, f'Audio Encoder: {model_info["audio_dim"]} ‚Üí 64', fontsize=10)
ax8.text(0.1, 0.6, f'Text Encoder: {model_info["text_dim"]} ‚Üí 128', fontsize=10)
ax8.text(0.1, 0.5, 'Attention Fusion', fontsize=10, style='italic')
ax8.text(0.1, 0.4, f'Brain Decoder: ‚Üí {model_info["brain_regions"]}', fontsize=10)
ax8.text(0.1, 0.3, f'Total Parameters: {model_info["total_parameters"]:,}', fontsize=10)
ax8.text(0.1, 0.2, f'Test R¬≤: {test_r2:.4f}', fontsize=12, fontweight='bold', color='red')
ax8.set_xlim(0, 1)
ax8.set_ylim(0, 1)
ax8.axis('off')

# 9. Feature importance heatmap
ax9 = plt.subplot(3, 4, 9)
# Compute feature importance as mean absolute response
feature_importance = np.mean(np.abs(predictions_np), axis=0)
top_50 = np.argsort(feature_importance)[-50:]
importance_matrix = feature_importance[top_50].reshape(10, 5)
im = plt.imshow(importance_matrix, cmap='viridis', aspect='auto')
plt.colorbar(im, ax=ax9)
plt.title('Top 50 Brain Regions Activity')
plt.xlabel('Region Index (reshaped)')
plt.ylabel('Region Index (reshaped)')

# 10. Error analysis
ax10 = plt.subplot(3, 4, 10)
errors = np.abs(predictions_np - targets_np)
mean_errors = np.mean(errors, axis=0)
plt.plot(mean_errors[:200], linewidth=1.5)
plt.xlabel('Brain Region')
plt.ylabel('Mean Absolute Error')
plt.title('Prediction Error by Brain Region')
plt.grid(True, alpha=0.3)

# 11. Cumulative explained variance
ax11 = plt.subplot(3, 4, 11)
# PCA on full data to get explained variance
pca_full = PCA()
pca_full.fit(predictions_np)
cumvar = np.cumsum(pca_full.explained_variance_ratio_)
plt.plot(cumvar[:100], linewidth=2)
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.grid(True, alpha=0.3)

# 12. Summary statistics
ax12 = plt.subplot(3, 4, 12)
stats_text = f"""Model Performance Summary

Test Loss: {test_loss:.4f}
Test R¬≤: {test_r2:.4f}

Dataset Statistics:
‚Ä¢ Training samples: {len(datasets['train'])}
‚Ä¢ Validation samples: {len(datasets['val'])}
‚Ä¢ Test samples: {len(datasets['test'])}

Model Statistics:
‚Ä¢ Parameters: {model_info['total_parameters']:,}
‚Ä¢ Brain regions: {model_info['brain_regions']}
‚Ä¢ Epochs trained: {len(history['train_loss'])}

Best Validation R¬≤: {max(history['val_r2']):.4f}
"""
ax12.text(0.05, 0.95, stats_text, fontsize=10, verticalalignment='top',
          fontfamily='monospace', bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgray"))
ax12.set_xlim(0, 1)
ax12.set_ylim(0, 1)
ax12.axis('off')

plt.tight_layout()
plt.show()

print("‚úÖ Visualizations complete!")

# ================================
# PART 9: ANALYSIS AND INSIGHTS
# ================================

print("\nüîç Analysis and Insights...")

# Compute detailed statistics
def analyze_predictions(targets, predictions):
    """Analyze prediction quality"""

    # Overall statistics
    mse = np.mean((targets - predictions) ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(targets - predictions))

    # Per-region R¬≤ scores
    region_r2 = []
    for i in range(targets.shape[1]):
        r2 = r2_score(targets[:, i], predictions[:, i])
        if not np.isnan(r2):
            region_r2.append(r2)

    region_r2 = np.array(region_r2)

    # Find best and worst performing regions
    best_regions = np.argsort(region_r2)[-10:]
    worst_regions = np.argsort(region_r2)[:10]

    return {
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'mean_r2': np.mean(region_r2),
        'std_r2': np.std(region_r2),
        'best_regions': best_regions,
        'worst_regions': worst_regions,
        'region_r2': region_r2
    }

# Analyze test predictions
analysis = analyze_predictions(targets_np, predictions_np)

print("üìä Detailed Analysis Results:")
print(f"   Mean Squared Error: {analysis['mse']:.6f}")
print(f"   Root Mean Squared Error: {analysis['rmse']:.6f}")
print(f"   Mean Absolute Error: {analysis['mae']:.6f}")
print(f"   Mean R¬≤ across regions: {analysis['mean_r2']:.4f} ¬± {analysis['std_r2']:.4f}")
print(f"   Best performing regions: {analysis['best_regions']}")
print(f"   Worst performing regions: {analysis['worst_regions']}")

# ================================
# PART 10: COMPARISON WITH REAL ALGONAUTS
# ================================

print("\nüèÜ How this compares to real Algonauts 2025:")

comparison_text = """
EDUCATIONAL DEMO vs REAL ALGONAUTS 2025 TRIBE:

üß† Model Architecture:
   Demo: Simplified 3-layer encoders + attention fusion
   Real: 1B parameter model with pre-trained foundation models

üìä Training Data:
   Demo: 2,000 synthetic samples
   Real: 80+ hours fMRI data per subject (Courtois NeuroMod)

üéØ Performance:
   Demo: R¬≤ ‚âà {:.3f} (synthetic data)
   Real: Won Algonauts 2025 competition (260+ teams)

‚öôÔ∏è Foundation Models:
   Demo: Simple linear encoders
   Real: V-JEPA 2, Wav2Vec2-BERT, Llama 3.2-3B

üé¨ Input Modalities:
   Demo: Random feature vectors
   Real: Movie frames, audio tracks, dialogue transcripts

üß™ Brain Responses:
   Demo: 1,000 synthetic brain regions
   Real: Actual fMRI from 1,000+ cortical regions

üíª Computational Requirements:
   Demo: Runs in Google Colab
   Real: Requires high-performance computing clusters
""".format(test_r2)

print(comparison_text)

# ================================
# PART 11: NEXT STEPS AND RESOURCES
# ================================

print("\nüöÄ Next Steps for Learning More:")

next_steps = """
1. üìö Study the Real Implementation:
   ‚Ä¢ Read the TRIBE paper: https://www.arxiv.org/abs/2507.22229
   ‚Ä¢ Explore the code: https://github.com/facebookresearch/algonauts-2025

2. üß† Learn about Brain Encoding:
   ‚Ä¢ Try BERG toolkit: https://github.com/gifale95/BERG
   ‚Ä¢ Explore Algonauts 2023 tutorial: github.com/gifale95/algonauts_2023

3. üéØ Work with Real Data:
   ‚Ä¢ Natural Scenes Dataset (NSD): naturalscenesdataset.org
   ‚Ä¢ Courtois NeuroMod: https://www.cneuromod.ca/

4. üõ†Ô∏è Advanced Tools:
   ‚Ä¢ Nilearn for fMRI analysis: nilearn.github.io
   ‚Ä¢ MNE for EEG/MEG: mne.tools

5. üèÜ Participate in Challenges:
   ‚Ä¢ Follow Algonauts Project: algonautsproject.com
   ‚Ä¢ Join brain modeling competitions
"""

print(next_steps)

print("\nüéâ Educational Demo Complete!")
print("You now understand the core concepts behind the Algonauts 2025 winning model!")

# Save final results
results_summary = {
    'model_info': model_info,
    'test_performance': {
        'loss': test_loss,
        'r2_score': test_r2
    },
    'training_history': history,
    'analysis': analysis
}

print(f"\nüíæ Results saved to memory. Test R¬≤ score: {test_r2:.4f}")
print("This notebook demonstrates the core concepts of multimodal brain encoding!")
```

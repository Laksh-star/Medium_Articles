```python
# Install required libraries
!pip install openai opencv-python tqdm matplotlib numpy

import cv2
import base64
import os
import numpy as np
from openai import OpenAI
from pathlib import Path
from google.colab import files
import getpass
from tqdm import tqdm
import matplotlib.pyplot as plt
import sys

# Step 1: Specify the MP4 file path or upload
VIDEO_PATH = input("Enter the path to your 26:40 'The Tramp' MP4 clip (e.g., /content/THE TRAMP (1915) -- Charlie Chaplin, Edna Purviance.mp4), or press Enter to upload: ")
if not VIDEO_PATH.strip():
    print("No path provided. Please upload your 'The Tramp' MP4 clip.")
    uploaded = files.upload()
    VIDEO_PATH = list(uploaded.keys())[0]

# Verify file exists
if not os.path.exists(VIDEO_PATH):
    raise ValueError(f"File not found at {VIDEO_PATH}. Run '!ls /content/' to check.")

# Get duration and FPS
vid = cv2.VideoCapture(VIDEO_PATH)
if not vid.isOpened():
    raise ValueError(f"Cannot open video file: {VIDEO_PATH}")
fps = vid.get(cv2.CAP_PROP_FPS)
frame_count = vid.get(cv2.CAP_PROP_FRAME_COUNT)
duration = frame_count / fps / 60  # Minutes
max_frame = int(frame_count)
vid.release()
print(f"Video duration: {duration:.2f} minutes, FPS: {fps}, Total frames: {frame_count}", flush=True)
if abs(duration - (26 + 40/60)) > 1:
    print("Warning: Duration differs from 26:40. Ranges will be scaled.", flush=True)

# Step 2: Securely input OpenAI API key
API_KEY = getpass.getpass("Enter your OpenAI API key: ")

# Configuration
FRAME_RATE = 2                   # 2 fps
OUTPUT_DIR = "frames"            # Directory for temporary frames

# Multiple-choice questions with ranges
MCQS = [
    {
        "question": "According to the opening title card, what year was 'The Tramp' copyrighted by Essanay Film Mfg. Co.?",
        "options": ["a) 1912", "b) 1915", "c) 1918"],
        "timestamp": "Opening title card",
        "frame_start": 0,
        "frame_end": 20
    },
    {
        "question": "What initially causes the Tramp (Charlie Chaplin) to intervene when the young woman is by the river/bushes?",
        "options": ["a) He hears her call for help.", "b) He sees two other men attacking her and trying to take her money/purse.", "c) Her father asks him to protect her."],
        "timestamp": "Early scene by river/bushes",
        "frame_start": 120,
        "frame_end": 360
    },
    {
        "question": "After saving the woman, the farmer offers the Tramp a reward. What is this reward?",
        "options": ["a) Money", "b) A place to sleep", "c) A job working on the farm"],
        "timestamp": "After saving the woman",
        "frame_start": 360,
        "frame_end": 720
    },
    {
        "question": "What task involving an animal does the Tramp attempt after being given work on the farm?",
        "options": ["a) Feeding chickens", "b) Shearing a sheep", "c) Milking a cow"],
        "timestamp": "Farm work scene",
        "frame_start": 1200,
        "frame_end": 1800
    },
    {
        "question": "What prompts the Tramp to finally leave the farm at the end of the clip?",
        "options": ["a) He is fired for bad farm work.", "b) He finds and reads a note written by the young woman.", "c) The robbers return and chase him away."],
        "timestamp": "End of the clip",
        "frame_start": 2900,
        "frame_end": 3200
    }
]

def extract_frames(video_path, output_dir, frame_rate, frame_ranges, max_frame):
    """Extract frames with verbose logging and RGB conversion."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    vid = cv2.VideoCapture(video_path)
    if not vid.isOpened():
        raise ValueError(f"Cannot open video file: {video_path}")

    frame_interval = int(vid.get(cv2.CAP_PROP_FPS) / frame_rate)
    frame_files = []
    extracted_frames = set()

    # Scale ranges if needed
    scale = min(1.0, max_frame / (26.67 * 60 * frame_rate))
    adjusted_ranges = [(max(0, int(start * scale)), min(int(end * scale), max_frame)) for start, end in frame_ranges]
    total_frames_to_extract = sum(min(end - start, int((end / frame_rate) - (start / frame_rate))) for start, end in adjusted_ranges)

    print(f"Expected to extract ~{total_frames_to_extract} frames.", flush=True)
    with tqdm(total=total_frames_to_extract, desc="Extracting frames", unit="frame") as pbar:
        for range_idx, (start, end) in enumerate(adjusted_ranges):
            print(f"Processing range {range_idx+1}: {start} to {end}...", flush=True)

            # Make sure we get at least 3 frames for each range
            min_frames_per_range = 3
            frames_from_range = 0

            vid.set(cv2.CAP_PROP_POS_FRAMES, start)
            count = start
            retries = 0
            while count < end:
                ret, frame = vid.read()
                if not ret:
                    if retries < 5:
                        retries += 1
                        vid.set(cv2.CAP_PROP_POS_FRAMES, count)
                        print(f"Retry {retries} for frame {count}...", flush=True)
                        continue
                    print(f"Failed at frame {count}. Stopping range.", flush=True)
                    break

                # Either extract at the regular interval OR force extract to meet minimum frames per range
                if (count % frame_interval == 0 and count not in extracted_frames) or \
                   (frames_from_range < min_frames_per_range and count % (frame_interval // 2) == 0 and count not in extracted_frames):
                    if frame is not None:
                        # Convert BGR to RGB
                        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                        # Apply basic enhancement for clearer images
                        # Increase contrast for better visibility (especially for old films)
                        frame = cv2.convertScaleAbs(frame, alpha=1.2, beta=10)

                        # Resize maintaining aspect ratio with padding
                        h, w = frame.shape[:2]
                        if h > w:
                            new_h, new_w = 512, int(512 * w / h)
                        else:
                            new_h, new_w = int(512 * h / w), 512

                        frame_resized = cv2.resize(frame, (new_w, new_h))

                        # Add black padding to make it square 512x512
                        square_img = np.zeros((512, 512, 3), dtype=np.uint8)
                        # Calculate center offset
                        x_offset = (512 - new_w) // 2
                        y_offset = (512 - new_h) // 2
                        # Place the resized image in the center
                        square_img[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = frame_resized

                        frame_path = os.path.join(output_dir, f"frame_{range_idx+1}_{count:04d}.jpg")
                        if cv2.imwrite(frame_path, cv2.cvtColor(square_img, cv2.COLOR_RGB2BGR), [int(cv2.IMWRITE_JPEG_QUALITY), 95]):
                            frame_files.append((count, frame_path, range_idx))
                            extracted_frames.add(count)
                            frames_from_range += 1
                            pbar.update(1)
                        else:
                            print(f"Failed to save frame {count}: Invalid or empty frame.", flush=True)
                count += 1
                retries = 0

                # If we've extracted enough frames from this range, we can move on
                if frames_from_range >= min_frames_per_range and count > (start + (end - start) // 3):
                    # Break after getting minimum frames AND we're at least 1/3 through the range
                    # This ensures we don't miss important content later in the range
                    break

    vid.release()
    print(f"Extracted {len(frame_files)} unique frames.", flush=True)

    # Preview frames for each range
    if frame_files:
        plt.figure(figsize=(15, 10))

        # Find one frame from each question range for preview
        for i in range(len(MCQS)):
            frames_for_q = [f for f in frame_files if f[2] == i]
            if frames_for_q:
                mid_idx = len(frames_for_q) // 2
                f_num, f_path, _ = frames_for_q[mid_idx]
                img = cv2.imread(f_path)
                if img is not None:
                    plt.subplot(2, 3, i + 1)
                    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                    plt.title(f"Frame ~{f_num} (Q{i + 1})")
                    plt.axis("off")
                else:
                    print(f"Preview frame {f_path} is invalid.", flush=True)

        plt.tight_layout()
        plt.show()

    return sorted(frame_files, key=lambda x: x[0])

def image_to_base64(image_path):
    """Convert image file to base64 string with validation."""
    try:
        file_size = os.path.getsize(image_path) / 1024  # KB
        if file_size > 500:
            print(f"Warning: Frame {image_path} is {file_size:.2f}KB, may be too large.", flush=True)

        # Ensure image exists and can be read
        img = cv2.imread(image_path)
        if img is None:
            print(f"Warning: Cannot read image at {image_path}", flush=True)
            return None

        # Resize to ensure it's not too large (optional)
        img = cv2.resize(img, (512, 512))
        _, buffer = cv2.imencode('.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])

        # Convert to base64
        encoded = base64.b64encode(buffer).decode("utf-8")
        print(f"Encoded {image_path}: {len(encoded)} bytes.", flush=True)
        return encoded
    except Exception as e:
        print(f"Error encoding {image_path}: {e}", flush=True)
        return None

def analyze_frames_with_gpt4(frame_files, mcqs):
    """Send frames to GPT-4.1 with updated API structure."""
    client = OpenAI(api_key=API_KEY)
    answers = []

    for mcq_idx, mcq in enumerate(tqdm(mcqs, desc="Analyzing questions", unit="question")):
        # Find frames specifically for this question range
        relevant_frames = [(f_num, f_path) for f_num, f_path, range_idx in frame_files if range_idx == mcq_idx]

        if not relevant_frames:
            answers.append(f"Q{mcq_idx + 1}: Cannot answer without valid frames")
            print(f"No frames for Q{mcq_idx + 1} ({mcq['timestamp']}).", flush=True)
            continue

        print(f"Q{mcq_idx + 1}: Found {len(relevant_frames)} frames: {[f_num for f_num, _ in relevant_frames]}", flush=True)

        # Use all available frames but limit to at most 3 to avoid API limitations
        best_frames = relevant_frames
        if len(best_frames) > 3:
            # Pick frames spread evenly throughout the range
            indices = [int(i * len(relevant_frames) / 3) for i in range(3)]
            best_frames = [relevant_frames[i] for i in indices]

        content_list = []

        # First add the text prompt
        prompt = f"Answer this question about a scene from Charlie Chaplin's 1915 silent film 'The Tramp'. The images show frames from {mcq['timestamp']}. These are black and white silent film frames from 1915, so they may be grainy or low contrast, but they are valid images. Answer in the format: 'Q{mcq_idx + 1}: [Option]'.\n\n"
        prompt += f"Q{mcq_idx + 1}: {mcq['question']}\n"
        prompt += "\n".join(mcq['options']) + "\n"

        content_list.append({"type": "input_text", "text": prompt})

        # Then add each image
        valid_images = 0
        for _, f_path in best_frames:
            try:
                # Verify the image before encoding
                img = cv2.imread(f_path)
                if img is None:
                    print(f"Warning: Cannot read image at {f_path} - skipping", flush=True)
                    continue

                # Convert to base64
                _, buffer = cv2.imencode('.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
                base64_img = base64.b64encode(buffer).decode("utf-8")

                content_list.append({
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_img}"
                })
                valid_images += 1
                print(f"Successfully added image {f_path} ({img.shape})", flush=True)
            except Exception as e:
                print(f"Error processing image {f_path}: {e}", flush=True)

        if valid_images == 0:
            answers.append(f"Q{mcq_idx + 1}: Cannot answer without valid frames")
            print(f"No valid images for Q{mcq_idx + 1}.", flush=True)
            continue

        print(f"Sending {valid_images} images for Q{mcq_idx + 1}...", flush=True)

        # Try a few different prompt variations if needed
        prompt_variations = [
            prompt,  # Original prompt
            f"Look at these frames from the 1915 Charlie Chaplin film 'The Tramp'. For Q{mcq_idx + 1}: {mcq['question']}\n{mcq['options'][0]}\n{mcq['options'][1]}\n{mcq['options'][2]}\nChoose the best option.",
            f"Based ONLY on the visible content in these film frames, answer: {mcq['question']}\n{mcq['options'][0]}\n{mcq['options'][1]}\n{mcq['options'][2]}"
        ]

        for variation_idx, prompt_text in enumerate(prompt_variations):
            if variation_idx > 0:
                # Replace the text prompt in content_list with the variation
                content_list[0] = {"type": "input_text", "text": prompt_text}
                print(f"Trying prompt variation {variation_idx}...", flush=True)

            retries = 0
            while retries < 3:
                try:
                    # Format the request exactly as in the working snippet
                    response = client.responses.create(
                        model="gpt-4o",
                        input=[
                            {
                                "role": "user",
                                "content": content_list
                            }
                        ]
                    )

                    response_text = response.output_text
                    print(f"Raw response: {response_text}", flush=True)

                    # Look for the answer pattern in different formats
                    if f"Q{mcq_idx + 1}:" in response_text:
                        answers.append(response_text)
                        print(f"Received response for Q{mcq_idx + 1}: {response_text}", flush=True)
                        break
                    elif any(opt.split(")")[0].lower() in response_text.lower() for opt in mcq["options"]):
                        # If the response contains a, b, or c but not formatted exactly as requested
                        for opt in mcq["options"]:
                            letter = opt.split(")")[0].lower()
                            if letter in response_text.lower():
                                formatted_response = f"Q{mcq_idx + 1}: {opt}"
                                answers.append(formatted_response)
                                print(f"Reformatted response: {formatted_response}", flush=True)
                                break
                        break
                    else:
                        print(f"Warning: Response format does not contain an answer choice: {response_text[:100]}...", flush=True)
                        if retries == 2 and variation_idx == len(prompt_variations) - 1:
                            answers.append(f"Q{mcq_idx + 1}: Could not determine answer from response")
                        retries += 1
                except Exception as e:
                    retries += 1
                    print(f"Error processing Q{mcq_idx + 1}, retry {retries}: {e}", flush=True)
                    if retries == 3 and variation_idx == len(prompt_variations) - 1:
                        answers.append(f"Q{mcq_idx + 1}: Cannot answer due to API error - {str(e)[:100]}")
                        print(f"Failed after all retries for Q{mcq_idx + 1}.", flush=True)

            # If we got a proper answer, break out of the prompt variations loop
            if answers and (f"Q{mcq_idx + 1}:" in answers[-1] or any(opt.split(")")[0].lower() in answers[-1].lower() for opt in mcq["options"])):
                break

    return answers

def clean_up_frames(output_dir):
    """Delete temporary frame files."""
    try:
        for file in os.listdir(output_dir):
            os.remove(os.path.join(output_dir, file))
        os.rmdir(output_dir)
        print("Cleaned up temporary frames.", flush=True)
    except Exception as e:
        print(f"Error cleaning frames: {e}", flush=True)

def main():
    try:
        # Adjust frame ranges
        frame_ranges = [(mcq["frame_start"], mcq["frame_end"]) for mcq in MCQS]

        # Extract frames
        print("Extracting frames for specific ranges...", flush=True)
        frame_files = extract_frames(VIDEO_PATH, OUTPUT_DIR, FRAME_RATE, frame_ranges, max_frame)
        if not frame_files:
            print("No frames extracted. Check video file.", flush=True)
            return

        # Analyze frames with GPT-4.1
        print("Analyzing frames with GPT-4.1...", flush=True)
        answers = analyze_frames_with_gpt4(frame_files, MCQS)

        # Print results
        print("\nAnalysis Results:", flush=True)
        for answer in answers:
            print(answer, flush=True)

        # Expected answers
        print("\nExpected Answers for 'The Tramp' (26:40 clip):", flush=True)
        print("Q1: b) 1915", flush=True)
        print("Q2: b) He sees two other men attacking her and trying to take her money/purse.", flush=True)
        print("Q3: c) A job working on the farm", flush=True)
        print("Q4: c) Milking a cow", flush=True)
        print("Q5: b) He finds and reads a note written by the young woman.", flush=True)
        print("Calculate accuracy (e.g., 4/5 correct = 80%) to validate GPT-4o's performance against the claim (~72% accuracy).", flush=True)

    except Exception as e:
        print(f"Error: {e}", flush=True)
    finally:
        # Clean up
        if os.path.exists(OUTPUT_DIR):
            clean_up_frames(OUTPUT_DIR)

if __name__ == "__main__":
    main()
```

```python
# Install required libraries
!pip install openai opencv-python tqdm matplotlib numpy

import cv2
import base64
import os
import numpy as np
from openai import OpenAI
from pathlib import Path
from google.colab import files
import getpass
from tqdm import tqdm
import matplotlib.pyplot as plt
import sys

# Step 1: Specify the MP4 file path or upload
VIDEO_PATH = input("Enter the path to your 26:40 'The Tramp' MP4 clip (e.g., /content/THE TRAMP (1915) -- Charlie Chaplin, Edna Purviance.mp4), or press Enter to upload: ")
if not VIDEO_PATH.strip():
    print("No path provided. Please upload your 'The Tramp' MP4 clip.")
    uploaded = files.upload()
    VIDEO_PATH = list(uploaded.keys())[0]

# Verify file exists
if not os.path.exists(VIDEO_PATH):
    raise ValueError(f"File not found at {VIDEO_PATH}. Run '!ls /content/' to check.")

# Get duration and FPS
vid = cv2.VideoCapture(VIDEO_PATH)
if not vid.isOpened():
    raise ValueError(f"Cannot open video file: {VIDEO_PATH}")
fps = vid.get(cv2.CAP_PROP_FPS)
frame_count = vid.get(cv2.CAP_PROP_FRAME_COUNT)
duration = frame_count / fps / 60  # Minutes
max_frame = int(frame_count)
vid.release()
print(f"Video duration: {duration:.2f} minutes, FPS: {fps}, Total frames: {frame_count}", flush=True)
if abs(duration - (26 + 40/60)) > 1:
    print("Warning: Duration differs from 26:40. Ranges will be scaled.", flush=True)

# Step 2: Securely input OpenAI API key
API_KEY = getpass.getpass("Enter your OpenAI API key: ")

# Configuration
FRAME_RATE = 2                   # 2 fps
OUTPUT_DIR = "frames"            # Directory for temporary frames

# Multiple-choice questions with ranges
MCQS = [
    {
        "question": "According to the opening title card, what year was 'The Tramp' copyrighted by Essanay Film Mfg. Co.?",
        "options": ["a) 1912", "b) 1915", "c) 1918"],
        "timestamp": "Opening title card",
        "frame_start": 0,
        "frame_end": 20
    },
    {
        "question": "What initially causes the Tramp (Charlie Chaplin) to intervene when the young woman is by the river/bushes?",
        "options": ["a) He hears her call for help.", "b) He sees two other men attacking her and trying to take her money/purse.", "c) Her father asks him to protect her."],
        "timestamp": "Early scene by river/bushes",
        "frame_start": 120,
        "frame_end": 360
    },
    {
        "question": "After saving the woman, the farmer offers the Tramp a reward. What is this reward?",
        "options": ["a) Money", "b) A place to sleep", "c) A job working on the farm"],
        "timestamp": "After saving the woman",
        "frame_start": 360,
        "frame_end": 720
    },
    {
        "question": "What task involving an animal does the Tramp attempt after being given work on the farm?",
        "options": ["a) Feeding chickens", "b) Shearing a sheep", "c) Milking a cow"],
        "timestamp": "Farm work scene",
        "frame_start": 1200,
        "frame_end": 1800
    },
    {
        "question": "What prompts the Tramp to finally leave the farm at the end of the clip?",
        "options": ["a) He is fired for bad farm work.", "b) He finds and reads a note written by the young woman.", "c) The robbers return and chase him away."],
        "timestamp": "End of the clip",
        "frame_start": 2900,
        "frame_end": 3200
    }
]

def extract_frames(video_path, output_dir, frame_rate, frame_ranges, max_frame):
    """Extract frames with verbose logging and RGB conversion."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    vid = cv2.VideoCapture(video_path)
    if not vid.isOpened():
        raise ValueError(f"Cannot open video file: {video_path}")

    frame_interval = int(vid.get(cv2.CAP_PROP_FPS) / frame_rate)
    frame_files = []
    extracted_frames = set()

    # Scale ranges if needed
    scale = min(1.0, max_frame / (26.67 * 60 * frame_rate))
    adjusted_ranges = [(max(0, int(start * scale)), min(int(end * scale), max_frame)) for start, end in frame_ranges]
    total_frames_to_extract = sum(min(end - start, int((end / frame_rate) - (start / frame_rate))) for start, end in adjusted_ranges)

    print(f"Expected to extract ~{total_frames_to_extract} frames.", flush=True)
    with tqdm(total=total_frames_to_extract, desc="Extracting frames", unit="frame") as pbar:
        for range_idx, (start, end) in enumerate(adjusted_ranges):
            print(f"Processing range {range_idx+1}: {start} to {end}...", flush=True)

            # Make sure we get at least 3 frames for each range
            min_frames_per_range = 3
            frames_from_range = 0

            vid.set(cv2.CAP_PROP_POS_FRAMES, start)
            count = start
            retries = 0
            while count < end:
                ret, frame = vid.read()
                if not ret:
                    if retries < 5:
                        retries += 1
                        vid.set(cv2.CAP_PROP_POS_FRAMES, count)
                        print(f"Retry {retries} for frame {count}...", flush=True)
                        continue
                    print(f"Failed at frame {count}. Stopping range.", flush=True)
                    break

                # Either extract at the regular interval OR force extract to meet minimum frames per range
                if (count % frame_interval == 0 and count not in extracted_frames) or \
                   (frames_from_range < min_frames_per_range and count % (frame_interval // 2) == 0 and count not in extracted_frames):
                    if frame is not None:
                        # Convert BGR to RGB
                        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                        # Apply basic enhancement for clearer images
                        # Increase contrast for better visibility (especially for old films)
                        frame = cv2.convertScaleAbs(frame, alpha=1.2, beta=10)

                        # Resize maintaining aspect ratio with padding
                        h, w = frame.shape[:2]
                        if h > w:
                            new_h, new_w = 512, int(512 * w / h)
                        else:
                            new_h, new_w = int(512 * h / w), 512

                        frame_resized = cv2.resize(frame, (new_w, new_h))

                        # Add black padding to make it square 512x512
                        square_img = np.zeros((512, 512, 3), dtype=np.uint8)
                        # Calculate center offset
                        x_offset = (512 - new_w) // 2
                        y_offset = (512 - new_h) // 2
                        # Place the resized image in the center
                        square_img[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = frame_resized

                        frame_path = os.path.join(output_dir, f"frame_{range_idx+1}_{count:04d}.jpg")
                        if cv2.imwrite(frame_path, cv2.cvtColor(square_img, cv2.COLOR_RGB2BGR), [int(cv2.IMWRITE_JPEG_QUALITY), 95]):
                            frame_files.append((count, frame_path, range_idx))
                            extracted_frames.add(count)
                            frames_from_range += 1
                            pbar.update(1)
                        else:
                            print(f"Failed to save frame {count}: Invalid or empty frame.", flush=True)
                count += 1
                retries = 0

                # If we've extracted enough frames from this range, we can move on
                if frames_from_range >= min_frames_per_range and count > (start + (end - start) // 3):
                    # Break after getting minimum frames AND we're at least 1/3 through the range
                    # This ensures we don't miss important content later in the range
                    break

    vid.release()
    print(f"Extracted {len(frame_files)} unique frames.", flush=True)

    # Preview frames for each range
    if frame_files:
        plt.figure(figsize=(15, 10))

        # Find one frame from each question range for preview
        for i in range(len(MCQS)):
            frames_for_q = [f for f in frame_files if f[2] == i]
            if frames_for_q:
                mid_idx = len(frames_for_q) // 2
                f_num, f_path, _ = frames_for_q[mid_idx]
                img = cv2.imread(f_path)
                if img is not None:
                    plt.subplot(2, 3, i + 1)
                    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                    plt.title(f"Frame ~{f_num} (Q{i + 1})")
                    plt.axis("off")
                else:
                    print(f"Preview frame {f_path} is invalid.", flush=True)

        plt.tight_layout()
        plt.show()

    return sorted(frame_files, key=lambda x: x[0])

def image_to_base64(image_path):
    """Convert image file to base64 string with validation."""
    try:
        file_size = os.path.getsize(image_path) / 1024  # KB
        if file_size > 500:
            print(f"Warning: Frame {image_path} is {file_size:.2f}KB, may be too large.", flush=True)

        # Ensure image exists and can be read
        img = cv2.imread(image_path)
        if img is None:
            print(f"Warning: Cannot read image at {image_path}", flush=True)
            return None

        # Resize to ensure it's not too large (optional)
        img = cv2.resize(img, (512, 512))
        _, buffer = cv2.imencode('.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])

        # Convert to base64
        encoded = base64.b64encode(buffer).decode("utf-8")
        print(f"Encoded {image_path}: {len(encoded)} bytes.", flush=True)
        return encoded
    except Exception as e:
        print(f"Error encoding {image_path}: {e}", flush=True)
        return None

def analyze_frames_with_gpt4(frame_files, mcqs):
    """Send frames to GPT-4.1 with updated API structure."""
    client = OpenAI(api_key=API_KEY)
    answers = []

    for mcq_idx, mcq in enumerate(tqdm(mcqs, desc="Analyzing questions", unit="question")):
        # Find frames specifically for this question range
        relevant_frames = [(f_num, f_path) for f_num, f_path, range_idx in frame_files if range_idx == mcq_idx]

        if not relevant_frames:
            answers.append(f"Q{mcq_idx + 1}: Cannot answer without valid frames")
            print(f"No frames for Q{mcq_idx + 1} ({mcq['timestamp']}).", flush=True)
            continue

        print(f"Q{mcq_idx + 1}: Found {len(relevant_frames)} frames: {[f_num for f_num, _ in relevant_frames]}", flush=True)

        # Use all available frames but limit to at most 3 to avoid API limitations
        best_frames = relevant_frames
        if len(best_frames) > 3:
            # Pick frames spread evenly throughout the range
            indices = [int(i * len(relevant_frames) / 3) for i in range(3)]
            best_frames = [relevant_frames[i] for i in indices]

        content_list = []

        # First add the text prompt
        prompt = f"Answer this question about a scene from Charlie Chaplin's 1915 silent film 'The Tramp'. The images show frames from {mcq['timestamp']}. These are black and white silent film frames from 1915, so they may be grainy or low contrast, but they are valid images. Answer in the format: 'Q{mcq_idx + 1}: [Option]'.\n\n"
        prompt += f"Q{mcq_idx + 1}: {mcq['question']}\n"
        prompt += "\n".join(mcq['options']) + "\n"

        content_list.append({"type": "input_text", "text": prompt})

        # Then add each image
        valid_images = 0
        for _, f_path in best_frames:
            try:
                # Verify the image before encoding
                img = cv2.imread(f_path)
                if img is None:
                    print(f"Warning: Cannot read image at {f_path} - skipping", flush=True)
                    continue

                # Convert to base64
                _, buffer = cv2.imencode('.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
                base64_img = base64.b64encode(buffer).decode("utf-8")

                content_list.append({
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_img}"
                })
                valid_images += 1
                print(f"Successfully added image {f_path} ({img.shape})", flush=True)
            except Exception as e:
                print(f"Error processing image {f_path}: {e}", flush=True)

        if valid_images == 0:
            answers.append(f"Q{mcq_idx + 1}: Cannot answer without valid frames")
            print(f"No valid images for Q{mcq_idx + 1}.", flush=True)
            continue

        print(f"Sending {valid_images} images for Q{mcq_idx + 1}...", flush=True)

        # Try a few different prompt variations if needed
        prompt_variations = [
            prompt,  # Original prompt
            f"Look at these frames from the 1915 Charlie Chaplin film 'The Tramp'. For Q{mcq_idx + 1}: {mcq['question']}\n{mcq['options'][0]}\n{mcq['options'][1]}\n{mcq['options'][2]}\nChoose the best option.",
            f"Based ONLY on the visible content in these film frames, answer: {mcq['question']}\n{mcq['options'][0]}\n{mcq['options'][1]}\n{mcq['options'][2]}"
        ]

        for variation_idx, prompt_text in enumerate(prompt_variations):
            if variation_idx > 0:
                # Replace the text prompt in content_list with the variation
                content_list[0] = {"type": "input_text", "text": prompt_text}
                print(f"Trying prompt variation {variation_idx}...", flush=True)

            retries = 0
            while retries < 3:
                try:
                    # Format the request exactly as in the working snippet
                    response = client.responses.create(
                        model="gpt-4.1-nano-2025-04-14",
                        input=[
                            {
                                "role": "user",
                                "content": content_list
                            }
                        ]
                    )

                    response_text = response.output_text
                    print(f"Raw response: {response_text}", flush=True)

                    # Look for the answer pattern in different formats
                    if f"Q{mcq_idx + 1}:" in response_text:
                        answers.append(response_text)
                        print(f"Received response for Q{mcq_idx + 1}: {response_text}", flush=True)
                        break
                    elif any(opt.split(")")[0].lower() in response_text.lower() for opt in mcq["options"]):
                        # If the response contains a, b, or c but not formatted exactly as requested
                        for opt in mcq["options"]:
                            letter = opt.split(")")[0].lower()
                            if letter in response_text.lower():
                                formatted_response = f"Q{mcq_idx + 1}: {opt}"
                                answers.append(formatted_response)
                                print(f"Reformatted response: {formatted_response}", flush=True)
                                break
                        break
                    else:
                        print(f"Warning: Response format does not contain an answer choice: {response_text[:100]}...", flush=True)
                        if retries == 2 and variation_idx == len(prompt_variations) - 1:
                            answers.append(f"Q{mcq_idx + 1}: Could not determine answer from response")
                        retries += 1
                except Exception as e:
                    retries += 1
                    print(f"Error processing Q{mcq_idx + 1}, retry {retries}: {e}", flush=True)
                    if retries == 3 and variation_idx == len(prompt_variations) - 1:
                        answers.append(f"Q{mcq_idx + 1}: Cannot answer due to API error - {str(e)[:100]}")
                        print(f"Failed after all retries for Q{mcq_idx + 1}.", flush=True)

            # If we got a proper answer, break out of the prompt variations loop
            if answers and (f"Q{mcq_idx + 1}:" in answers[-1] or any(opt.split(")")[0].lower() in answers[-1].lower() for opt in mcq["options"])):
                break

    return answers

def clean_up_frames(output_dir):
    """Delete temporary frame files."""
    try:
        for file in os.listdir(output_dir):
            os.remove(os.path.join(output_dir, file))
        os.rmdir(output_dir)
        print("Cleaned up temporary frames.", flush=True)
    except Exception as e:
        print(f"Error cleaning frames: {e}", flush=True)

def main():
    try:
        # Adjust frame ranges
        frame_ranges = [(mcq["frame_start"], mcq["frame_end"]) for mcq in MCQS]

        # Extract frames
        print("Extracting frames for specific ranges...", flush=True)
        frame_files = extract_frames(VIDEO_PATH, OUTPUT_DIR, FRAME_RATE, frame_ranges, max_frame)
        if not frame_files:
            print("No frames extracted. Check video file.", flush=True)
            return

        # Analyze frames with GPT-4.1
        print("Analyzing frames with GPT-4.1...", flush=True)
        answers = analyze_frames_with_gpt4(frame_files, MCQS)

        # Print results
        print("\nAnalysis Results:", flush=True)
        for answer in answers:
            print(answer, flush=True)

        # Expected answers
        print("\nExpected Answers for 'The Tramp' (26:40 clip):", flush=True)
        print("Q1: b) 1915", flush=True)
        print("Q2: b) He sees two other men attacking her and trying to take her money/purse.", flush=True)
        print("Q3: c) A job working on the farm", flush=True)
        print("Q4: c) Milking a cow", flush=True)
        print("Q5: b) He finds and reads a note written by the young woman.", flush=True)
        print("Calculate accuracy (e.g., 4/5 correct = 80%) to validate GPT-4o's performance against the claim (~72% accuracy).", flush=True)

    except Exception as e:
        print(f"Error: {e}", flush=True)
    finally:
        # Clean up
        if os.path.exists(OUTPUT_DIR):
            clean_up_frames(OUTPUT_DIR)

if __name__ == "__main__":
    main()
```

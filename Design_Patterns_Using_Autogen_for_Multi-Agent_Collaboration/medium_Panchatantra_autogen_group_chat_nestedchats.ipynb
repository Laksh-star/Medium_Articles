```python
llm_config = {"model": "gpt-3.5-turbo"}
```

```python
import autogen
from autogen import ConversableAgent, AssistantAgent
```

```python


senior_strategist = ConversableAgent(
    name="Senior_Strategist",
    system_message="As an experienced strategist, you excel at anticipating market trends and"
        "competitive dynamics. Your ability to see the big picture and plan for"
        "contingencies is crucial for the startup's success.",
    description="Devise a winning strategy to maintain the startup\'s edge",
    llm_config=llm_config,
    human_input_mode="ALWAYS",
)

agility_specialist = ConversableAgent(
    name="Agility_Specialist",
    system_message="You have a keen eye for agility and execution, you're adept at finding ways"
        "for the startup to move faster than the competition. Your role is to"
        "optimize processes and empower the team to act swiftly.",
    llm_config=llm_config,
    description="Ensure the startup can respond quickly to the competition",
    human_input_mode="NEVER",
)
market_analyst = ConversableAgent(
    name="Market_Analyst",
    system_message="As a diligent market analyst, you're tasked with keeping a close eye"
        "on the competitive landscape. Your mission is to identify potential"
        "blind spots, challenge assumptions, and ensure the startup remains adaptable.",
    description="Provide vigilant market monitoring and insights",
    llm_config=llm_config,
    
    human_input_mode="NEVER",
)
```

```python
from autogen import GroupChat
group_chat_with_introductions = GroupChat(
    agents=[senior_strategist, agility_specialist, market_analyst],
    messages=[],
    max_round=3,
    send_introductions=True,
)
```

```python
from autogen import GroupChatManager
group_chat_manager_with_intros = GroupChatManager(
    groupchat=group_chat_with_introductions,
    llm_config=llm_config,
)
```

```python
nested_chats = [
    {
        "recipient": group_chat_manager_with_intros,
        "summary_method": "reflection_with_llm",
        "summary_prompt": "Summarize the sequence of operations used to turn " "into a solution.",
    },
    {
        "recipient": agility_specialist,
        "message": "A detailed 2-page note with recommended action",
        "summary_method": "reflection_with_llm",
    },
    {
        "recipient": senior_strategist,
        "message": "A three paragraph recommendations report on enhancing the startup agility",
        "max_turns": 1,
        "summary_method": "last_msg",
    },
]
```

```python
senior_strategist.register_nested_chats(
    nested_chats,
    # The trigger function is used to determine if the agent should start the nested chat
    # given the sender agent.
    # In this case, the agent will not start the nested chats if the sender is
    # from the nested chats' recipient to avoid recursive calls.
    trigger=lambda sender: sender not in [group_chat_manager_with_intros],
)
```

```python
# Instead of using `initiate_chat` method to start another conversation,
# we can use the `generate_reply` method to get single reply to a message directly.
reply = senior_strategist.generate_reply(
    messages=[{"role": "user", "content": "Unable to control the employee turnover. What to do?"}]
)
```
